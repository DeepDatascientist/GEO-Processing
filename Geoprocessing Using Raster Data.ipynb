```python
# importing modules

import numpy as np
import pandas as pd
import rasterio
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
```

```python
# Importing and plotting citi shape files

cities = gpd.read_file('belgian_cities.shp')
cities.plot(cmap='jet',column='NAME_4',figsize=(10,10))
```

```python
# Plotting Area of Intrest Shape

AOI = gpd.read_file('area_of_interest_.shp')
AOI.plot()
```

```python
# Displaying both shapefiles Together

fig, ax = plt.subplots(1)
cities.plot(ax=ax,cmap='jet')#using cmap to enhance the visuals
AOI.plot(ax=ax)

```

```python
# Intersecting(It has intersected belgium cities into AOI arc)

cities_in_AOI = gpd.overlay(cities,AOI,how='intersection')
cities_in_AOI.plot(figsize=(10,10),cmap= 'jet',column= 'NAME_4')
```

```python
# Assigning new column- Area

cities_in_AOI['Area(km2)']=cities_in_AOI.area/1000000
```
